````python
# 分担
target_dir = "###/" # sad, angry, happy, neutralのどれか
class_dir = "###/" # train, test, validationのどれか
sketched_list_txt = "###.txt" # <生成済み画像リスト>.txt
````

## （３）CLIPassoの実行

・以下実行でスケッチ画像生成スタート.  
・（注意点）12時間？でセッションが切れるので現状は再度初めから実行とTODO実施が必要

（以下はコメント）  
・ストローク数は32本で固定。  
・16,32で実行時間大きく変化しない、16では人間主観で顔と認識できるスケッチがうまく生成できなかった。画像サイズは48✖️48だと荒すぎたので150✖️150にしました。  
・seedを2000に固定してしまって実行時間を短くしています。  
・実行すると指定したディレクトリ配下のjpgファイル全てにスケッチ画像を生成し始めます。元画像数はtrainデータだと各クラス4000枚ぐらい。  
・1枚生成するのに7分ぐらい必要、1セッション(12時間として)およそ100枚生成できる。     
・生成したマッピングやスケッチを確認したい場合は"/content/CLIPasso/output_sketches"配下にある。  
・データセット用の最終的なスケッチ画像は"/content/drive/Shareddrives/SketchWorldModels/sketch_datasets"配下に保存されていく。  

````python
import glob
import cv2
import cairosvg
import time
import pickle

%cd /content/CLIPasso
%matplotlib inline

mask_object = 0 
fix_scale = 0 
num_strokes =  32

input_image_dir = '/content/drive/Shareddrives/SketchWorldModels/plain_datasets/' + class_dir + target_dir
output_image_dir = '/content/drive/Shareddrives/SketchWorldModels/sketch_datasets/' + class_dir + target_dir # output_dirはrunで使用しているので注意
sketched_list_dir = '/content/drive/Shareddrives/SketchWorldModels/' + sketched_list_txt

files = glob.glob(input_image_dir + '*')

# runtimeエラーがあるので処理ごとにロードする
f = open(sketched_list_dir, 'rb')
prev_sketched_list = pickle.load(f)
sketched_set = set(prev_sketched_list)

for f in files:
    file_name = f.split('/')[-1]
    rm_ext_file_name = file_name.split('.')[0]
    target_image = f

    if rm_ext_file_name in sketched_set:
      continue

    # runtimeエラーがあるので処理ごとにロードする
    f = open(sketched_list_dir, 'rb')
    prev_sketched_list = pickle.load(f)
    sketched_set = set(prev_sketched_list)

    %run start_sketching.py --target_file $target_image --num_sketches 3 -colab --mask_object $mask_object --fix_scale $fix_scale --num_strokes $num_strokes
    output_file_dir = "/content/CLIPasso/output_sketches/" + class_dir + target_dir + rm_ext_file_name
    
    # svgからpngに変換
    cairosvg.svg2png(url=output_file_dir + "/best.svg", write_to=output_file_dir + "/best.png")

    time.sleep(10) # pngが生成されるまで待機

    # 150*150にリサイズと、RGBA形式に注意してpngからjpgに変換
    img = cv2.imread(output_file_dir + "/best.png", -1)
    img48 = cv2.resize(img, dsize=(150, 150))
    a = img48[:, :, 3]
    for row in range(len(a)):
      for col in range(len(a[row])):
        if a[row][col] >= 128: a[row][col] = 0
        else: a[row][col] = 255

    cv2.imwrite(output_image_dir + rm_ext_file_name + '_sketched.jpg', a, [cv2.IMWRITE_JPEG_QUALITY, 100])

    print("Save " + rm_ext_file_name + "_sketched.jpg to '" + output_image_dir + "'")
    
    sketched_set.add(rm_ext_file_name)
    sketched_list = list(sketched_set)
    print(sketched_list)
    f = open(sketched_list_dir,'wb')
    pickle.dump(sketched_set, f)
````
